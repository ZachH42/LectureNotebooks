## Color Spaces

In this notebook 
* HSV colors
* Switching color-spaces
* Manipulating colors

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
lake_img = cv2.imread(os.path.relpath('../../Data/New_Zealand_Lake.jpg'), cv2.IMREAD_COLOR)
```

```python
# We can split the colors using the cv2.split command.  This gives use 3 monochromatic images
# We can merge to get the original image back

B,G,R = cv2.split(lake_img)   
# Remember, openCV used BGR (not RGB) order

lake_img_rgb = cv2.merge([R,G,B])

# Show the images
plt.figure(figsize=[18, 5])

plt.subplot(151);plt.imshow(lake_img);  plt.title("Original-BGR")
plt.subplot(152);plt.imshow(B,cmap="gray");plt.title("Blue")
plt.subplot(153);plt.imshow(G,cmap="gray");plt.title("Green")
plt.subplot(154);plt.imshow(R,cmap="gray");plt.title("Red")
plt.subplot(155);plt.imshow(lake_img_rgb);plt.title("Merged-RGB")

```

## cvtColor
This function changes the color scheme of the image

The color change command has the format
* img_new = cv2.cvtColor(img,cv2.COLOR_{X}2{Y})
where {X} and {Y} are the abreviations below

The (main) avaliable color schemes are:
* BGR  - Blue/Green/Red     [Default for openCV]
* BGRA - Blue/Green/Red + Alpha channel
* RGB  - Red/Green/Blue 
* RGBA - Red/Green/Blue + Alpha channel
* GRAY - Grayscale (monochromatic)
* HSV - Hue Saturation Value

```python
# We could have done the same BGR 2 RBG conversion like this
lake_img_rgb = cv2.cvtColor(lake_img,cv2.COLOR_BGR2RGB)

# or convert to HSV
lake_img_hsv = cv2.cvtColor(lake_img,cv2.COLOR_BGR2HSV)
# SPlit the HSV
H,S,V = cv2.split(lake_img_hsv)

# Show the images
plt.figure(figsize=[18, 5])

plt.subplot(141);plt.imshow(lake_img_rgb);  plt.title("Original-RBG")
plt.subplot(142);plt.imshow(H,cmap="gray");plt.title("Hue")
plt.subplot(143);plt.imshow(S,cmap="gray");plt.title("Saturation")
plt.subplot(144);plt.imshow(V,cmap="gray");plt.title("Value")
```

## Modifying colors
If we want to modify an image, we need to 
* First convert to the color scheme in which we will modify
* Modify the image in the correct third dimension entry
* Convert to RGB so the matplotlib.pyplot function will work

```python
lake_img_hsv = cv2.cvtColor(lake_img,cv2.COLOR_BGR2HSV)

lake_img_hsv[:,:,0]+=50 

lake_img_mod = cv2.cvtColor(lake_img_hsv,cv2.COLOR_HSV2RGB)

# Show the images
plt.figure(figsize=[18, 5])

plt.subplot(121);plt.imshow(lake_img_rgb);  plt.title("Original-RBG")
plt.subplot(122);plt.imshow(lake_img_mod); plt.title("Increased hue by 50")
```
